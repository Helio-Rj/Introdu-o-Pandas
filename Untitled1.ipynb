```python
import pandas as pd

#  from IPython.core.display_functions import display
#  se quizer usar o 'display'

# importar a base de dados
tabela_vendas = pd.read_excel('Vendas.xlsx')

# visualizar a base de dados
pd.set_option('display.max_columns', None)

#  Exibição de lojas/quantidade
lojas = tabela_vendas[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()

#  Adicionando uma nova coluna a tabela com valor zerado
lojas.loc[:, 'ID'] = 0
#  Adicionando uma nova coluna a tabela com valor zerado
lojas.loc[:, 'Código Venda'] = 0

#  Adicionando uma nova coluna a tabela com valor zerado
lojas.loc[:, 'Produto'] = 0
#  Adicionando uma nova coluna a tabela com valor zerado
lojas.loc[:, 'Valor Unitário'] = 0
#  Adicionando uma nova coluna a tabela com valor zerado
lojas.loc[:, 'Valor Final'] = 0
#  Adicionando uma nova coluna a tabela com valor zerado
lojas.loc[:, 'Comissão'] = 0
#  Adicionando uma nova coluna a tabela com valor zerado
lojas.loc[:, 'Imposto'] = 0
```

```python
lojas = lojas[['ID', 'Produto', 'Código Venda', 'Quantidade', 'Valor Unitário', 'Valor Final', 'Comissão', 'Imposto']] 
display(lojas)
```
